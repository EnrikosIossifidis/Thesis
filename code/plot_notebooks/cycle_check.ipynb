```python
def get_cycle(a,cur):
    b = []
    for i in a:
        curix = np.where(cur==i)
        wlast = curix[:-1]
        if len(wlast) == 0:
            curlen = len(cur)
            lastix = curix[0]
            newix = (lastix+1)%curlen
            newval = cur[newix]
        else:
            curlen = len(cur[wlast][0])
            lastix = curix[-1:][0]
            newix = (lastix+1)%curlen
            newval = cur[wlast][0][newix]    
        b.append(newval[0])
    return b

def compute_cycle(c1,c2,a):
    print(a)
    b = get_cycle(a,np.array(c1))
    c = get_cycle(b,np.array(c2))
    print(c)
    valid_comb = np.all((a==np.array(c)) == False)
    print(valid_comb)
```

```python
for c1 in cycles:
    for c2 in cycles:
        print(c1,c2)
        compute_cycle(c1,c2,a)
```
