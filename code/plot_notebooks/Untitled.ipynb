```python
d = pd.DataFrame(data={})
d['xs'] = np.linspace(0, 5, 15)
d['ys'] = np.linspace(0, 5, 15)

X, Y = np.meshgrid(d['xs'], d["ys"])
data = np.array([X,Y]).reshape(2, -1).T

d2 = pd.DataFrame(data={})
d2['X'] = data[:,0]
d2['Y'] = data[:,1]

d2['I'] = d2.apply(lambda row: np.random.random(),axis=1)
d2['WMS'] = d2.apply(lambda row: int(np.random.random()*10),axis=1)
sns.scatterplot(data=d2,x='X',y='Y',size='I',hue='WMS')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
